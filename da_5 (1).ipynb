```python
import seaborn as sns
iris = sns.load_dataset("iris")
```

```python
iris

#species열을 기준으로 각 그룹별 열의 평균
iris.groupby('species').mean()
```

```python
def p2p_ratio(x):
    return(x.max() / x.min())
```

```python
#각각의 붓꽃 종류별 최대값과 최소값을 구하여 비율(최대값/최소값) 출력
iris.groupby('species').agg(p2p_ratio)
```

```python
iris.groupby('species').describe().T
```

```python
#transform함수:그룹별로 계산 -> 데이터프레임에 결과 반영
#꽃잎 길이가 해당 종에서 대/중/소 중에서 어디에 속하는지?
```

```python
def q3cut(data):#data에는 각 그룹별 petal_length 데이터가 전달됨
    return (pd.qcut(data, 3, labels=['소','중','대']))    
```

```python
#iris -> 종별 그룹화 -> petal_length -> 3개 구간(소/중/대, qcut함수)
iris['petal_length_class']=iris.groupby(['species']).petal_length.transform(q3cut)
```

```python
#iris의 setosa 종에 대한 petal_length_class의 종류별 데이터 건수
iris.groupby('species')['petal_length_class'].value_counts()['setosa']
```

```python
#각 그룹에서 최대팁-최저팁 
#성별,흡연자
```

```python
def ptp(x):
    return x.max()-x.min()
```

```python
tips.groupby(['sex','smoker'])[['tip']].mean()
tips.groupby(['sex','smoker'])[['tip']].max()

tips.groupby(['sex','smoker'])[['tip']].agg(ptp)
```

```python
tips.groupby(['sex','smoker'])[['tip']].agg([ptp,'mean','max','min'])
tips.groupby(['sex','smoker']).agg([ptp,'mean','max','min'])[['tip']]
```

```python
tips.groupby(['sex','smoker'])[['total_bill']].agg([ptp,'mean','max','min'])
```
