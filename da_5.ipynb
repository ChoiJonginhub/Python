```python
# 어느 회사의 전반기(1월 ~ 6월) 실적을 나타내는 데이터프레임과 
# 후반기(7월 ~ 12월) 실적을 나타내는 데이터프레임을 만든 뒤 합친다. 
# 실적 정보는 “매출”, “비용”, “이익” 으로 이루어진다. (이익 = 매출 - 비용).
# 또한 1년간의 총 실적을 마지막 행으로 덧붙인다.
```

```python
df1=pd.DataFrame({
    '연도': ['1월','2월','3월','4월','5월','6월'],
    '매출': [1000,2000,4500,1500,4600,2000],
    '비용': [800,1000,4000,1000,2000,4000],
}, columns=['연도','매출','비용'])
df1
```

```python
df2=pd.DataFrame({
    '연도': ['7월','8월','9월','10월','11월','12월'],
    '매출': [1000,2500,4000,4000,5000,7000],
    '비용': [1800,1500,4000,3000,4000,5000],
}, columns=['연도','매출','비용'])
df2
```

```python
df1['이익']=df1['매출']-df1['비용']
df2['이익']=df2['매출']-df2['비용']
```

```python
df2
```

```python
df3=pd.concat([df1,df2])
df3
```

```python


df3.iloc[-1] = 
['총실적',df3['매출'].sum(),df3['비용'].sum(),df3['매출'].sum()-df3['비용'].sum()]
```
