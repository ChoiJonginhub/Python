```python
#연속형 -> 카테고리 변환 : cut, qcut함수
ages = [0, 1, 20, 30, 23, 37, 31, 61, 20, 41, 32, 101]
bins = [1, 20, 30, 50, 70, 100] #카테고리 기준값

#1<미성년자<=20, (1,20] ...
labels = ["미성년자", "청년", "중년", "장년", "노년"]
#범위 외 데이터 -> NaN으로 취급됨
c=pd.cut(ages, bins, labels=labels)
c
```

```python
c.categories
c.codes
```

```python
ages
```

```python
df4=pd.DataFrame(ages, columns=['ages'])
```

```python
df4
#df4의 ages을 범주형으로
df4['age_cate']=pd.cut(df4.ages,bins,labels=labels)
```

```python
#데이터 병합 : 공통된 열 인덱스 또는 행 인덱스를 기준으로 합침
```

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
```

```python
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
pd.merge(df1,df2) # pd.merge(df1,df2,how='inner') 와 동일, 디폴트, 내부조인
#내부조인?공통으로 존재하는 키(열 이름) 기준으로 합침

#pd.merge(df1,df2,left_on='고객번호', right_on='고객번호2')
```

```python
pd.merge(df1,df2,how='outer') #외부조인, outer, 키 값이 어느 한쪽에만 있더라도 병합됨
```

```python
pd.merge(df1,df2,how='left') 
#df1(left)을 기준으로 df2와 merge, df1에는 모든 자료가 merge가 됨
pd.merge(df1,df2,how='right') 
#df2(right)을 기준으로 df1과 merge, df2에는 모든 자료가 merge가 됨
```
