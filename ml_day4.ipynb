```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
boston=load_boston()
dfx=pd.DataFrame(boston.data, columns=boston.feature_names)
dfy=pd.DataFrame(boston.target, columns=['MEDV']) #집값
```

```python
dfx
```

```python
dfy
```

```python
from sklearn.model_selection import train_test_split
#선형회귀
from sklearn.linear_model import LinearRegression
#MSE
from sklearn.metrics import mean_squared_error
#from statsmodels.formula.api import ols
#선형회귀모델=ols(종속변수~독립변수, 데이터).fit()
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(dfx, dfy, test_size=0.2, random_state=415)
```

```python
model=LinearRegression() #선형회귀모델 객체 생성
model.fit(xtrain, ytrain) #학습
```

```python
#h(x) = w1*x1 + w2*x2 + ... + w13*x13 + b
#가중치 확인
model.coef_
#h(x) = -1.08*x1 + 6.89*x2 + ... + -5.56*x13 + 45.567
model.intercept_
```

```python
#선형회귀모델을 이용하여 테스트 데이터를 예측
xtest
```

```python
pred=model.predict(xtest)
pred
```

```python
ytest
```

```python
mse=mean_squared_error(ytest, pred)
```

```python
mse
```
