```python
import re
text=["abcdefg","abcde","abc"]
r=re.compile("abc+")
for i in text:
    if r.match(i):
        print("Match")
    else:
        print("Skip")
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import re
driver=webdriver.Chrome(ChromeDriverManager().install())
url="https://news.naver.com/main/read.naver?mode=LSD&mid=shm&sid1=103&oid=052&aid=0001697611"
driver.get(url)
html=driver.page_source
soup=BeautifulSoup(html, 'html.parser')
t=soup.select("#articleBodyContents")
temp=re.findall("[영][상|하][ \d*.]+[도]",str(t))
snow=re.findall("[\d]?.[\d]cm", str(t))
tel= re.findall("\d{2,3}-\d{3,4}-\d{4}", str(t))
email= re.findall("[\w]+@[\w]+\..+", str(t))
print("온도", temp)
print("적설량", snow)
print("연락처", tel)
print("이메일", email)
```
