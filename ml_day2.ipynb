```python
"""
SVM : 결정 경계, 각 클러스터로부터 최대한 멀리 떨어져 있는 결정 경계가 좋다
SV(Support Vectors) : 결정 경계와 가장 가깝게 있는 데이터들을 의미 -> 결정 경계를 정의하는 역할
마진 : 결정 경계와 서포트 벡터 사이의 거리

하이퍼 퍼러미터
C(cost):비용, 결정 경계의 마진을 결정하는 퍼러미터
C 작으면 -> soft margin(결정경계와 서포트벡터 사이의 거리가 커짐), 오류 허용(언더피팅)
C 커지면 -> hard margin(결정경계와 서포트벡터 사이의 거리가 작아짐), 오류 허용 x(오버피팅)

gamma:결정경계선의 유연함
gamma 작아지면 -> 직선에 가까워짐, 언더피팅
gamma 커지면 -> 꾸불꾸불해짐, 오버피팅
"""
```
